```python
from numpy import array

def mat22(a11,a12,a21,a22):
    return a11*a22 - a12*a21

def mat33(a11,a12,a13,a21,a22,a23,a31,a32,a33):
    return a13*a21*a32 + a11*a22*a33 + a12*a23*a31 - a12*a21*a33 - a13*a22*a31 - a11*a23*a32

filename = 'test_data.txt'

# 이 알고리즘의
# 2X2 행렬 입력 조건
# a11 a12
# a21 a22
# b1
# b2

# 이 알고리즘의
# 3X3 행렬 입력 조건
# a11 a12 a13
# a21 a22 a23
# a31 a32 a33
# b1
# b2
# b3

check = 0
test = []
wfilename = open('out_data.txt','w',encoding = 'ascii')

with open(filename, encoding = 'ascii') as file_object:
    lines = file_object.readlines()

Mat = []
Mat2 = Mat
Mat3 = [0 for i in range(6)]

for line in lines :
    Mat.append(line.split())

for idx_i, val_i in enumerate(Mat):
    for idx_j,val_j in enumerate(val_i):
        Mat2[idx_i][idx_j] = int(Mat[idx_i][idx_j])

checkM = 0
tmplen = len(Mat2)

for i in range(0,tmplen):
    if(len(Mat2[i]) != 0):
        Mat3[check] = Mat2[i]
        if(check == 0):
            checkM = len(Mat3[check])
        check = check + 1
        
        if(checkM == 2 and i == 5):
            break
        
        if(checkM == 3 and i == 7):
            break
        
print(Mat3)

if len(Mat3[0]) == 2 :
    
    det = mat22(Mat3[0][0],Mat3[0][1],Mat3[1][0],Mat3[1][1])
    det1 = mat22(Mat3[2][0],Mat3[3][0],Mat3[0][1],Mat3[1][1])
    det2 = mat22(Mat3[0][0],Mat3[2][0],Mat3[1][0],Mat3[3][0])
    
    x1 = det1/det
    x2 = det2/det
    
    print("x1 : ",x1,"(",det1,"/",det,")")
    print("x2 : ",x2,"(",det2,"/",det,")")
    
    wfilename.write("x1 : " + str(x1) + " (" + str(det1) + "/" + str(det) + ")"+ "\n")
    wfilename.write("x2 : " + str(x2) + " (" + str(det2) + "/" + str(det) + ")")
    
elif len(Mat3[0]) == 3 :
    det = mat33(Mat3[0][0],Mat3[0][1],Mat3[0][2],Mat3[1][0],Mat3[1][1],Mat3[1][2],Mat3[2][0],Mat3[2][1],Mat3[2][2])
    det1 = mat33(Mat3[3][0],Mat3[0][1],Mat3[0][2],Mat3[4][0],Mat3[1][1],Mat3[1][2],Mat3[5][0],Mat3[2][1],Mat3[2][2])
    det2 = mat33(Mat3[0][0],Mat3[3][0],Mat3[0][2],Mat3[1][0],Mat3[4][0],Mat3[1][2],Mat3[2][0],Mat3[5][0],Mat3[2][2])
    det3 = mat33(Mat3[0][0],Mat3[0][1],Mat3[3][0],Mat3[1][0],Mat3[1][1],Mat3[4][0],Mat3[2][0],Mat3[2][1],Mat3[5][0])
    
    x1 = det1/det
    x2 = det2/det
    x3 = det3/det

    print("x1 : ",x1,"(",det1,"/",det,")")
    print("x2 : ",x2,"(",det2,"/",det,")")
    print("x3 : ",x3,"(",det3,"/",det,")")
    
    wfilename.write("x1 : " + str(x1) + " (" + str(det1) + "/" + str(det) + ")"+ "\n")
    wfilename.write("x2 : " + str(x2) + " (" + str(det2) + "/" + str(det) + ")"+ "\n")
    wfilename.write("x3 : " + str(x3) + " (" + str(det3) + "/" + str(det) + ")")
```
